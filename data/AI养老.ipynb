```python
import pandas as pd
import random
import numpy as np
from faker import Faker
from datetime import datetime

fake = Faker("zh_CN")
random.seed(42)
np.random.seed(42)

n = 500
rows = []

# åŸå¸‚ã€èŒä¸šç­‰å‚è€ƒåˆ—è¡¨
cities = ["åŒ—äº¬","ä¸Šæµ·","å¹¿å·","æ·±åœ³","æ­å·","æˆéƒ½","æ­¦æ±‰","è¥¿å®‰","å—äº¬","å¦é—¨",
          "é’å²›","å¤©æ´¥","é•¿æ²™","é‡åº†","è‹å·","å®æ³¢","éƒ‘å·","ç¦å·","åˆè‚¥","å¤§è¿"]
occupations = ["ç¨‹åºå‘˜","æ•™å¸ˆ","åŒ»ç”Ÿ","å¾‹å¸ˆ","è‡ªç”±èŒä¸š","å…¬åŠ¡å‘˜","ä¼ä¸šç®¡ç†","é‡‘èåˆ†æå¸ˆ",
               "å¸‚åœºè¥é”€","åˆ¶é€ ä¸šå·¥äºº","ç§‘ç ”äººå‘˜","ä¸ªä½“å•†æˆ·","å­¦ç”Ÿ","é€€ä¼‘äººå‘˜"]
educations = ["é«˜ä¸­","å¤§ä¸“","æœ¬ç§‘","ç¡•å£«","åšå£«"]
maritals = ["æœªå©š","å·²å©š","ç¦»å¼‚","ä¸§å¶"]
risk_pref = ["ä¿å®ˆ","ç¨³å¥","å¹³è¡¡","ç§¯æ","æ¿€è¿›"]
health_status = ["ä¼˜","è‰¯","ä¸€èˆ¬","å·®"]

for i in range(n):
    # å¹´é¾„ï¼š18~65ï¼ŒæŒ‰ç°å®å¹´é¾„æ®µåŠ æƒ
    age_group = random.choices(
        population=[(18,25),(26,35),(36,45),(46,55),(56,65)],
        weights=[0.15,0.3,0.25,0.2,0.1]
    )[0]
    age = random.randint(age_group[0], age_group[1])
    birth_year = datetime.today().year - age
    birth = fake.date_of_birth(tzinfo=None, minimum_age=age, maximum_age=age)

    # æ”¶å…¥ä¸æ”¯å‡ºé€»è¾‘
    salary = round(np.random.normal(loc=8000 + age*200, scale=3000),2)
    salary = max(2000, min(salary, 60000))
    business_income = round(random.uniform(0, 20000 if age>25 else 5000),2)
    passive_income = round(random.uniform(0, 10000 if age>30 else 2000),2)
    inflow = salary + business_income + passive_income
    outflow = round(inflow * random.uniform(0.6, 1.1), 2)
    savings_rate = round((inflow - outflow)/inflow, 2)

    # æ¶ˆè´¹ç»†åˆ†åˆ†é…ï¼ˆæ€»å’Œâ‰ˆoutflowï¼‰
    # è®¾ç½®ä¸€ä¸ªåŸºç¡€åˆ†å¸ƒ
    consumption_keys = [
        "é¤é¥®","è¡£ç‰©","ä½æˆ¿","äº¤é€š","å¨±ä¹","æ•™è‚²åŸ¹è®­","åŒ»ç–—ä¿å¥","å¥èº«è¿åŠ¨","æ—…è¡Œåº¦å‡",
        "æ•°å­—äº§å“","å® ç‰©","å›¾ä¹¦å½±éŸ³","ç¾å®¹æŠ¤è‚¤","çº¿ä¸Šè´­ç‰©","çº¿ä¸‹è´­ç‰©","å¥¢ä¾ˆå“",
        "å®¶åº­æ—¥ç”¨å“","æ¯å©´","ç»¿è‰²ç¯ä¿","æ…ˆå–„æèµ "
    ]
    base_weights = np.random.dirichlet(np.ones(len(consumption_keys)),size=1)[0]
    consumption_detail = {k: round(outflow*w,2) for k,w in zip(consumption_keys, base_weights)}

    # èµ„äº§è´Ÿå€º
    asset_liquid = round(random.uniform(0, 300000 + age*10000), 2)
    asset_fixed = round(random.uniform(0, 1000000 + age*50000), 2)
    stock_value = round(random.uniform(0, 500000), 2)
    real_estate = round(random.uniform(0, 3000000 if age>30 else 500000), 2)
    total_asset = asset_liquid + asset_fixed + stock_value + real_estate

    debt_credit = round(random.uniform(0, 50000 + age*2000), 2)
    debt_mortgage = round(random.uniform(0, 1500000 if age>28 else 200000), 2)
    debt_other = round(random.uniform(0, 200000), 2)
    total_debt = debt_credit + debt_mortgage + debt_other
    net_asset = round(total_asset - total_debt, 2)
    debt_ratio = round(total_debt/total_asset, 2) if total_asset>0 else 0

    # ç¤¾ä¿ä¸ä¿éšœ
    social_security = round(random.uniform(0, 200000 if age>22 else 0), 2)
    pension_years = random.randint(0, max(0, age-22))
    housing_fund = round(random.uniform(0, 150000 if age>25 else 0), 2)
    insurance_premium = round(random.uniform(0, 20000 + age*300), 2)
    insurance_coverage = round(random.uniform(0, 2000000), 2)

    # ç”¨æˆ·åé¦ˆä¸ç›®æ ‡
    retire_target = random.randint(50,70)
    pension_goal = round(random.uniform(500000, 5000000), 2)
    exp_return_min = round(random.uniform(0.01, 0.06), 3)
    exp_return_max = round(random.uniform(0.06, 0.12), 3)

    # å¹³å°äº¤äº’
    platform_visits = random.randint(0, 30 if age<50 else 20)
    adoption_rate = round(random.uniform(0.1,0.9), 2)
    qa_count = random.randint(0, 40 if age<40 else 20)
    personalization_count = random.randint(0, 15)
    feedback_pos = round(random.uniform(0.3, 1.0), 2)
    trust_score = round(random.uniform(30, 95), 2)

    row = {
        "ç”¨æˆ·ID": f"U{str(i+1).zfill(6)}",
        "å¹´é¾„": age,
        "æ€§åˆ«": random.choice(["ç”·","å¥³"]),
        "æ‰€åœ¨åŸå¸‚": random.choice(cities),
        "èŒä¸š": random.choice(occupations),
        "æ•™è‚²ç¨‹åº¦": random.choice(educations),
        "å©šå§»çŠ¶å†µ": random.choice(maritals),
        "æœˆå·¥èµ„æ”¶å…¥": salary,
        "ç»è¥æ€§æ”¶å…¥": business_income,
        "è¢«åŠ¨æ”¶å…¥": passive_income,
        "æœˆæ€»æµå…¥": inflow,
        "æœˆæ€»æµå‡º": outflow,
        "å‚¨è“„ç‡": savings_rate,
        "æ´»æœŸå­˜æ¬¾": asset_liquid,
        "ç†è´¢äº§å“": asset_fixed,
        "è‚¡ç¥¨åŸºé‡‘": stock_value,
        "æˆ¿äº§ä¼°å€¼": real_estate,
        "æ€»èµ„äº§": total_asset,
        "å‡€èµ„äº§": net_asset,
        "ä¿¡ç”¨å¡æ¬ æ¬¾": debt_credit,
        "æˆ¿è´·ä½™é¢": debt_mortgage,
        "å…¶ä»–è´·æ¬¾": debt_other,
        "æ€»è´Ÿå€º": total_debt,
        "è´Ÿå€ºç‡": debt_ratio,
        "å…»è€é‡‘è´¦æˆ·ä½™é¢": social_security,
        "ç¼´çº³å¹´é™": pension_years,
        "ä½æˆ¿å…¬ç§¯é‡‘ä½™é¢": housing_fund,
        "å•†ä¸šä¿é™©å¹´ç¼´": insurance_premium,
        "ä¿é™©ä¿é¢": insurance_coverage,
        "é£é™©åå¥½": random.choice(risk_pref),
        "è®¡åˆ’é€€ä¼‘å¹´é¾„": retire_target,
        "ç›®æ ‡å…»è€é‡‘": pension_goal,
        "æœŸæœ›æ”¶ç›Šç‡ä¸‹é™": exp_return_min,
        "æœŸæœ›æ”¶ç›Šç‡ä¸Šé™": exp_return_max,
        "å¥åº·çŠ¶å†µ": random.choice(health_status),
        "å¹³å°æœˆè®¿é—®æ¬¡æ•°": platform_visits,
        "ç­–ç•¥é‡‡çº³ç‡": adoption_rate,
        "äº¤äº’é—®ç­”æ¬¡æ•°": qa_count,
        "ä¸ªæ€§åŒ–è®¾ç½®æ¬¡æ•°": personalization_count,
        "åé¦ˆç§¯æåº¦": feedback_pos,
        "ä¿¡ä»»è¯„åˆ†": trust_score,
    }
    # åŠ å…¥æ¶ˆè´¹ç»†åˆ†
    for k,v in consumption_detail.items():
        row[f"{k}æ¶ˆè´¹"] = v

    rows.append(row)

# ç”ŸæˆDataFrameå¹¶å¯¼å‡º
df = pd.DataFrame(rows)
df.to_csv("pension_mock_500.csv", index=False, encoding="utf-8-sig")
print("âœ… å·²ç”Ÿæˆ pension_mock_500.csv (500æ¡å®Œæ•´æ¨¡æ‹Ÿæ•°æ®)")
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from ipywidgets import interact, IntSlider, Dropdown

# 1ï¸âƒ£ è¯»å–æ¨¡æ‹Ÿæ•°æ®
df = pd.read_csv("pension_mock_500.csv")

# ======= åŸºç¡€å¤„ç† =======
# æŒ‰å¹´é¾„æ®µåˆ†ç»„
bins = [18, 25, 35, 45, 55, 65, 80]
labels = ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
df["å¹´é¾„æ®µ"] = pd.cut(df["å¹´é¾„"], bins=bins, labels=labels, right=False)

# æå–æ¶ˆè´¹ç±»åˆ—
consumption_cols = [c for c in df.columns if "æ¶ˆè´¹" in c and "æ€»" not in c]

# ======= å¯è§†åŒ–1ï¼šåŠ¨æ€å¹´é¾„æ®µæ”¶å…¥åˆ†å¸ƒ =======
def show_income_by_age(age_group):
    d = df[df["å¹´é¾„æ®µ"]==age_group]
    fig = px.histogram(d, x="æœˆæ€»æµå…¥", nbins=20,
                       title=f"ğŸ’°æœˆæ€»æµå…¥åˆ†å¸ƒï¼ˆå¹´é¾„æ®µï¼š{age_group}ï¼‰",
                       color_discrete_sequence=["steelblue"])
    fig.show()

interact(show_income_by_age, age_group=Dropdown(options=labels, description="å¹´é¾„æ®µ"))

# ======= å¯è§†åŒ–2ï¼šå‚¨è“„ç‡ vs å¹³å°ä¿¡ä»»åº¦ï¼ˆæ°”æ³¡å›¾ï¼‰ =======
fig = px.scatter(df, x="å‚¨è“„ç‡", y="å¹³å°ä¿¡ä»»è¯„åˆ†",
                 size="å¹³å°æœˆè®¿é—®æ¬¡æ•°", color="å¹´é¾„æ®µ",
                 hover_name="ç”¨æˆ·ID",
                 title="ğŸ’¡å‚¨è“„ç‡ vs å¹³å°ä¿¡ä»»åº¦ï¼ˆæ°”æ³¡å¤§å°=æœˆè®¿é—®æ¬¡æ•°ï¼‰")
fig.show()

# ======= å¯è§†åŒ–3ï¼šTopæ¶ˆè´¹ç±»åˆ«åŠ¨æ€é€‰æ‹© =======
def show_top_consumption(top_n=10):
    mean_consumption = df[consumption_cols].mean().sort_values(ascending=False).head(top_n)
    fig = px.bar(mean_consumption, x=mean_consumption.index, y=mean_consumption.values,
                 color=mean_consumption.index,
                 title=f"ğŸ½ï¸å¹³å‡æ¶ˆè´¹Top{top_n}")
    fig.update_layout(xaxis_title="æ¶ˆè´¹ç±»åˆ«", yaxis_title="å¹³å‡æœˆæ¶ˆè´¹(RMB)")
    fig.show()

interact(show_top_consumption, top_n=IntSlider(value=10, min=5, max=20))

# ======= å¯è§†åŒ–4ï¼šå¹³å°æœˆè®¿é—®æ¬¡æ•°æ—¶é—´åºåˆ—è¶‹åŠ¿æ¨¡æ‹Ÿ =======
# æ¨¡æ‹Ÿæ¯ä¸ªç”¨æˆ·12ä¸ªæœˆçš„è®¿é—®æ¬¡æ•°
user_sample = df.sample(1, random_state=42).iloc[0]
time_series = pd.DataFrame({
    "æœˆä»½": range(1,13),
    "è®¿é—®æ¬¡æ•°": np.random.poisson(lam=user_sample["å¹³å°æœˆè®¿é—®æ¬¡æ•°"], size=12)
})
fig = px.line(time_series, x="æœˆä»½", y="è®¿é—®æ¬¡æ•°",
              title=f"ğŸ“ˆç”¨æˆ· {user_sample['ç”¨æˆ·ID']} å¹³å°è®¿é—®æ¬¡æ•°è¶‹åŠ¿")
fig.show()

# ======= å¯è§†åŒ–5ï¼šå¹´é¾„ vs æ”¶å…¥ vs æ¶ˆè´¹ä¸‰ç»´æ°”æ³¡ =======
fig = px.scatter_3d(df,
    x="å¹´é¾„", y="æœˆæ€»æµå…¥", z="æœˆæ€»æµå‡º",
    color="å¹´é¾„æ®µ", size="å‚¨è“„ç‡",
    title="ğŸ”µ å¹´é¾„-æ”¶å…¥-æ”¯å‡º ä¸‰ç»´å…³ç³»",
    hover_data=["å¹³å°ä¿¡ä»»è¯„åˆ†"]
)
fig.show()
```
